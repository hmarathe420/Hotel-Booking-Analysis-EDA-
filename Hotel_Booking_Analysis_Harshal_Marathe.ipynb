#### Missing Values/Null Values

````python
# Missing Values/Null Values Count
df.isnull().sum()
````

````python
# Visualizing the missing values
num_of_nan = []
hotel_name = []
for i in df.columns:
  nan=df[i].isnull().sum()
  num_of_nan.append(nan)
  hotel_name.append(i)

plt.figure(figsize = (25,10))
srn.barplot(x = hotel_name, y = num_of_nan)
plt.plot()
````

````python
# visualizing null values with another method by using missingno library
import missingno as mns
mns.bar(df)
plt.plot()
````

````python
# visualizing null values with matrix 
mns.matrix(df)
plt.plot()
````
