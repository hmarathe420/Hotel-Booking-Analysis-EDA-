## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
df.isnull().sum()
````

````python
# handling missing values in agent column
df['agent'].replace(np.nan,0, inplace = True)
````

````python
# Handling missing values in company column
df['company'].replace(np.nan,0, inplace = True)
````

````python
# removing null values
df.dropna(inplace = True)
````

````python
#checking whether there is any null values remaining 
df.isnull().sum()
````

````python
#checking on which days booking was done more, whether on weekend or weekdays
df[['stays_in_weekend_nights','stays_in_week_nights']].sum()
````

````python
#checking which hotel was booked more city hotels or Resort hotel
hotel_counts= df['hotel'].value_counts()
print(hotel_counts)
````

````python
# calculating the number of city hotels was canceled
city_hotel_df =  df[df['hotel'] == 'City Hotel']
city_hotel_df = city_hotel_df[city_hotel_df['is_canceled'] == 1]
len(city_hotel_df)
````

````python
#calculating the number of resort hotel was canceled
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel = resort_hotel[resort_hotel['is_canceled'] == 1]
len(resort_hotel)
````

````python
#calculating the yearly booking
year = df['arrival_date_year'].unique()
booking = [len(df[df['arrival_date_year'] == 2015]),len(df[df['arrival_date_year'] == 2016]),len(df[df['arrival_date_year'] == 2017])]
yearly_booking = dict(zip(year, booking))
print(yearly_booking)
````

````python
#getting in which year city hotel booking is more.
City_hotel_yearly_booking = [len(df[df['hotel'] == 'City Hotel'][df[df['hotel'] == 'City Hotel']['arrival_date_year']==2015]) \
                             ,len(df[df['hotel'] == 'City Hotel'][df[df['hotel'] == 'City Hotel']['arrival_date_year']==2016])\
                             ,len(df[df['hotel'] == 'City Hotel'][df[df['hotel'] == 'City Hotel']['arrival_date_year']==2017])]
City_hotel_yearly_booking = dict(zip(list(yearly_booking.keys()),City_hotel_yearly_booking))
print(City_hotel_yearly_booking)
````

````python
#getting in which year resort hotel booking is more.
Resort_hotel_yearly_booking = [len(df[df['hotel']== 'Resort Hotel'][df[df['hotel']== 'Resort Hotel']['arrival_date_year']==2015]),\
                               len(df[df['hotel']== 'Resort Hotel'][df[df['hotel']== 'Resort Hotel']['arrival_date_year']==2016]),\
                               len(df[df['hotel']== 'Resort Hotel'][df[df['hotel']== 'Resort Hotel']['arrival_date_year']==2017])]
Resort_hotel_yearly_booking = dict(zip(yearly_booking.keys(),Resort_hotel_yearly_booking))
Resort_hotel_yearly_booking
````

````python
#getting week number list from dataframe.
week_number_list = sorted(list(df['arrival_date_week_number'].unique()))
print(week_number_list)
````

````python
# calculation number of cancelation on weekly basis
cancelation_list_over_week_numbers = []
for num in week_number_list:
  temp = df[df['arrival_date_week_number'] == num]
  cancel = len(temp[temp['is_canceled'] == 1])
  cancelation_list_over_week_numbers.append(cancel)

print(cancelation_list_over_week_numbers)
````

````python
#constructing weekly cancelation dictionory
weekly_cancelation = dict(zip(week_number_list,cancelation_list_over_week_numbers))
print(weekly_cancelation)
````

````python
#calculation the number of booking over customer type.
customer_type_list = list(df['customer_type'].unique())
num_of_booking_over_customer_type = []
for name in customer_type_list:
  num_of_booking_over_customer_type.append(len(df[df['customer_type'] == name]))

print(customer_type_list)
print(num_of_booking_over_customer_type)
````

````python
#Numerical columns extraction
feature = []
for column in df.columns:
  if type(df[column][0]) == np.int64 or type(df[column][0]) == np.float64:
    feature.append(column)

new_df = df[feature]
new_df.head()
````
